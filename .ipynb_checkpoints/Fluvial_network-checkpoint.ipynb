```python
(mg, z) = read_esri_ascii("rastert_extract1.txt", name="topographic__elevation")
```

```python
min_z = np.min(z[np.where(z > 0)])
max_z = np.max(z[np.where(z > 0)])
#help(imshow_grid)
imshow_grid(mg, 'topographic__elevation', limits=(min_z, max_z))
```
